```python
import matplotlib.pyplot as plt
import os
import random
import cv2
```

```python
train_empty = '/content/gdrive/MyDrive/AIL_project/traffic/Final Dataset/training/Empty'
train_low = '/content/gdrive/MyDrive/AIL_project/traffic/Final Dataset/training/Low'
train_medium = '/content/gdrive/MyDrive/AIL_project/traffic/Final Dataset/training/Medium'
train_high = '/content/gdrive/MyDrive/AIL_project/traffic/Final Dataset/training/High'
train_jam = '/content/gdrive/MyDrive/AIL_project/traffic/Final Dataset/training/Traffic Jam'
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

# Empty

```python
images = os.listdir(train_empty)
img_paths = [os.path.join(train_empty, i) for i in images]

img_paths[:5]
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15,6))
for a in ax.flatten():
    img = random.choice(img_paths)
    image = cv2.imread(img)
    a.imshow(image)
```
